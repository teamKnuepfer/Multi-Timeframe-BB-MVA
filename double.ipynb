# Multi timeframe bollinger bands / moving average 

Running a simple backtest on new trade idea. Enter mean-reversion trades on the slower timeframe bollinger bands. Cover the position when price converges to the fast moving average.

# Parameters

```python
symbol = "NQM22"
slow = 30
fast = 5

tick_s = 4
tick_v = 0.5
multiplier = tick_s * tick_v

start = "20220314"

print(f"Symbol: {symbol} \nSlow Timeframe: {slow} minutes \nFast Timeframe: {fast} minutes")

print(f"\nEntry Signal: Slow timeframe bollinger bands \nExit Signal: Fast timeframe moving average")

print(f"\nStart Date: {start} \nEnd Date: Today")
```

```python
dates = [x.strftime("%Y%m%d") for x in pd.date_range(start=start, end=datetime.today(), freq='d')]
dates
```

```python
# get all trades in date range
trades = pd.DataFrame()
for i in dates[:-1]:
    trades = trades.append(get_trades(symbol, start=i))
trades
```

```python
try:
    trades = trades.drop(columns=["Exchange"])
except:
    pass
trades['ts'] = trades.Timestamp.apply(lambda x:x[0:16])
trades
```

```python
# get slow timeframe bars
slow = get_mins("NQM22",start="20220311",end="20220330",interval=slow)
slow = get_bbands(slow)
slow = slow.loc[slow.Date.apply(lambda x:datetime.strptime(x,"%Y-%m-%d").strftime("%Y%m%d")) >= start]
slow
```

```python
# get fast timeframe bars
fast = get_mins("NQM22",start="20220311",end="20220330",interval=fast)
fast = get_bbands(fast)
fast = fast.loc[fast.Date.apply(lambda x:datetime.strptime(x,"%Y-%m-%d").strftime("%Y%m%d")) >= start]
fast
```

```python
# merge dataframes
trades = pd.merge(trades,slow[["Timestamp","BBL","BBU"]],how='left',left_on='ts',right_on='Timestamp', suffixes=('','_DROP'))
trades = pd.merge(trades,fast[["Timestamp","SMA"]],how='left',left_on='ts',right_on='Timestamp', suffixes=('_slow','_fast'))
trades = trades.drop(columns=['Timestamp_DROP','Timestamp_fast']).rename(columns={'Timestamp_slow':'Timestamp'})
trades = trades.ffill()
trades
```

```python
# only trade during trading hours
day = trades.ts.apply(lambda x:x[-5:])
trades = trades.loc[(day >= "08:00") & (day <= "14:00")].reset_index(drop=True)
trades['Date'] = trades.Timestamp.apply(lambda x:x[0:10])
trades
```

```python
## creating signal column which determines when we are long/short
 
# determine initial entry points, which is when price is above or below slow bollinger bands
trades["Signal"] = [1 if x < lower else (-1 if x > upper else 0) for x,lower,upper in zip(trades.Price,trades.BBL,trades.BBU)]

# fill in signal columns until the point where we receive an exit signal (determined by "exit_signal" variable)
arr = trades.to_numpy()
for i in range(1,len(arr)):
    if arr[i-1][7] != arr[i][7]:
        arr[i][8] = 0
    elif arr[i-1][8] == -1:
        if arr[i][1] > arr[i][6]:
            arr[i][8] = -1
    elif arr[i-1][8] == 1:
        if arr[i][1] < arr[i][6]:
            arr[i][8] = 1
```

```python
df = pd.DataFrame(arr, columns=trades.columns)
df
```

```python
df['Delta'] = df.Price.diff().shift(-1)
df['Ticks'] = df.Delta * df.Signal
df['pnl'] = df.Ticks * multiplier
```

```python
dailypnl = df.groupby("Date").Pnl.sum()
pnl = df.groupby("ts").Pnl.sum()
```

```python
import plotly.graph_objects as go

fig = go.Figure(go.Scatter(x=dailypnl.index, y= dailypnl.cumsum()))
fig.update_layout(title = "Daily Cumulative PNL", height=600)
```

```python
fig = go.Figure(go.Scatter(x=pnl.index, y= pnl.cumsum()))
fig.update_layout(title = "Continuous Cumulative PNL", height = 600)
```

# Trade Analytics

```python
strades = df[df['Signal'] == 1].groupby((df['Signal'] != 1).cumsum()).agg({"pnl":"sum"}).reset_index(drop=True)
strades["start"] = df[df['Signal'] == 1].groupby((df['Signal'] != 1).cumsum()).agg({"Timestamp":"min"}).reset_index(drop=True)
strades["end"] = df[df['Signal'] == 1].groupby((df['Signal'] != 1).cumsum()).agg({"Timestamp":"max"}).reset_index(drop=True)
strades["duration (mins)"] = (pd.to_datetime(strades.end) - pd.to_datetime(strades.start)).apply(lambda x: x.total_seconds()/60)
strades['side'] = "Short"
strades = strades[["side","pnl",'duration (mins)',"start","end"]]
strades
```

```python
btrades = df[df['Signal'] == -1].groupby((df['Signal'] != -1).cumsum()).agg({"pnl":"sum"}).reset_index(drop=True)
btrades["start"] = df[df['Signal'] == -1].groupby((df['Signal'] != -1).cumsum()).agg({"Timestamp":"min"}).reset_index(drop=True)
btrades["end"] = df[df['Signal'] == -1].groupby((df['Signal'] != -1).cumsum()).agg({"Timestamp":"max"}).reset_index(drop=True)
btrades["duration (mins)"] = (pd.to_datetime(btrades.end) - pd.to_datetime(btrades.start)).apply(lambda x: x.total_seconds()/60)
btrades['side'] = "Long"
btrades = btrades[["side","pnl",'duration (mins)',"start","end"]]
btrades
```

```python
all = strades.append(btrades)

print("All Trade Statistics")
display(all.pnl.describe())
print("\n")

print("Long Trade Statistics")
display(btrades.pnl.describe())
print("\n")

print("Short Trade Statistics")
display(strades.pnl.describe())
```

```python
# P&L
print(f'Total pnl: ${df.pnl.sum():.2f}')
print(f'Short pnl: ${strades.pnl.sum():.2f}')
print(f'Long pnl: ${btrades.pnl.sum():.2f}')

```

```python
#create barchart of trade pnl
all = all.sort_values("start")

fig = go.Figure()
fig.add_trace(go.Bar(x=all.start, y= all.pnl))
fig.update_xaxes(type='category')
fig.update_layout(title = "All Trades", height = 700)
fig.show()

fig = go.Figure()
fig.add_trace(go.Bar(x=btrades.start, y= btrades.pnl))
fig.update_xaxes(type='category')
fig.update_layout(title = "Long Trades", height = 600)
fig.show()

fig = go.Figure()
fig.add_trace(go.Bar(x=strades.start, y= strades.pnl))
fig.update_xaxes(type='category')
fig.update_layout(title = "Short Trades", height = 600)
fig.show()
```
